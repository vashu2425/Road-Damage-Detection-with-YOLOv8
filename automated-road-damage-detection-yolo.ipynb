## Step 1: Import Required Libraries

```python
from ultralytics import YOLO
import torch
import cv2
import matplotlib.pyplot as plt
import os
```

## Step 2: Load and Configure the Model

```python
# Load a larger model
model = YOLO('yolov8m.pt')
```

## Step 3: Train the Model

```python
# Train the model with adjusted settings
model.train(
    data='/kaggle/input/datasetnew/dataset/road_damage.yaml',  
    epochs=100,                         # Train for more epochs to allow the model to converge
    imgsz=640,                          # Input image size
    batch=16,                           # Increased batch size
    name='road_damage_detection',      
    patience=5,                        # Allow more patience before stopping
    optimizer='AdamW',                  # Use AdamW optimizer for better convergence
    lr0=0.0001,                         # Lower initial learning rate
    augment=True,                       # Enable augmentation for better generalization
    mosaic=True,                        # Add mosaic augmentation
    mixup=0.3,                          # Increase mixup augmentation
    dropout=0.2                         # Regularization to prevent overfitting
)
```

## Step 4: Evaluate the Model

```python
metrics = model.val(data='/kaggle/input/datasetnew/dataset/road_damage.yaml')  # Evaluate the model
print(metrics)
```

## Step 5: Make Predictions

```python
results = model.predict(source='/kaggle/input/nddata/test', save=True, conf=0.5)
```

## Step 6: Final Evaluation Metrics

```python
metrics = model.val(data='/kaggle/input/datasetnew/dataset/road_damage.yaml')
print("Final Evaluation Metrics:")
print(f"Precision: {metrics['precision']}")
print(f"Recall: {metrics['recall']}")
print(f"mAP@0.5: {metrics['map50']}")
print(f"mAP@0.5:0.95: {metrics['map']}")
```

```python
# Clear GPU memory
torch.cuda.empty_cache()

# Load the trained YOLOv8 model (best.pt) and force to CPU (if GPU memory is insufficient)
model = YOLO('/kaggle/input/bestpoints/best (1).pt').to('cpu')

# Load a test image
img_path = '/kaggle/input/nddata/test/India_000380.jpg'
img = cv2.imread(img_path)

# Optionally resize image to reduce memory usage
img_resized = cv2.resize(img, (640, 640))  # Resize to 640x640 (or adjust as needed)

# Perform prediction
results = model(img_resized)

# Access the first result in the list
result = results[0]

# Access the prediction boxes and labels
boxes = result.boxes  # The bounding boxes predicted by YOLO
labels = boxes.cls  # The class labels predicted
confidences = boxes.conf  # The confidence scores of the predictions
xywh = boxes.xywh  # The bounding box coordinates in xywh format (center_x, center_y, width, height)

# Print the results in a simple format (without using pandas)
for i in range(len(boxes)):
    print(f"Prediction {i+1}:")
    print(f"  Label: {labels[i].item()}")
    print(f"  Confidence: {confidences[i].item()}")
    print(f"  Coordinates (x_center, y_center, width, height): {xywh[i].cpu().numpy()}")
    print("-" * 30)

# Define the output path (make sure the directory exists)
output_path = '/kaggle/working/result_image.jpg'

# Save the image with predictions
result.save(output_path)

# Check if the image is saved correctly
if os.path.exists(output_path):
    # Create a larger figure for displaying the image
    plt.figure(figsize=(12, 12))  # Adjust the size as needed (12x12 is an example)
    
    # Display the saved image with predictions using matplotlib
    img_with_predictions = cv2.imread(output_path)
    plt.imshow(cv2.cvtColor(img_with_predictions, cv2.COLOR_BGR2RGB))
    plt.axis('off')  # Hide axes
    plt.show()
else:
    print(f"Error: Image was not saved to {output_path}")
```
